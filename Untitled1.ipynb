```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
%matplotlib inline
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 50)
```

```python
from matplotlib.colors import ListedColormap

sns.set_palette("Set2")
cmap = ListedColormap(sns.color_palette("Set2")) 

pal = sns.color_palette("Set2") 
color_list = pal.as_hex() 

set2_green = color_list[0]
set2_orange = color_list[1]
set2_blue = color_list[2]
set2_rose = color_list[3]
set2_green2 = color_list[4]
set2_yellow = color_list[5]
set2_beige = color_list[6]
set2_grey = color_list[7]

color_list
```

```python
dataset = pd.read_csv("marketing_data.csv")

dataset.head(5)
```

```python
dataset.info()
```

```python
dataset.columns = dataset.columns.str.replace(' ', '')
```

```python
dataset.dt_customer = pd.to_datetime(dataset.dt_customer) 
dataset.income = pd.to_numeric(dataset.income.replace('[^0-9\.-]', '', regex=True))
```

```python
dataset.income.fillna(dataset.income.mean(),inplace=True)
```

```python
dataset.describe()
```

```python
median_year_birth = dataset[(dataset['year_birth'] >= 1933)]['year_birth'].median()
print(median_year_birth)
dataset['year_birth'].replace(to_replace = dataset[(dataset['year_birth'] <= 1933)]['year_birth'].tolist(), value = median_year_birth, inplace = True)
```

```python
dataset['age'] = pd.Timestamp('now').year - dataset['year_birth'] 

dataset['minors_home'] = dataset.kidhome + dataset.teenhome

mnt_columns = [column for column in dataset.columns if 'mnt' in column]
dataset['mnt_total'] =  dataset[mnt_columns].sum(axis=1)

purchases_columns = [column for column in dataset.columns if 'purchases' in column]
dataset['purchases_total'] = dataset[purchases_columns].sum(axis=1)

campaigns_column = [column for column in dataset.columns if 'cmp' in column] + ['response']
dataset['accepted_cmp_total'] = dataset[campaigns_column].sum(axis=1)
```

```python
dataset['education'].unique()
```

```python
dataset['education'].replace(to_replace = ['2n Cycle'], value = 'Master', inplace = True)
```

```python
dataset.marital_status.value_counts()
```

```python
dataset['marital_status'].replace(to_replace = ['Alone', 'YOLO', 'Absurd'], value = 'Single', inplace = True)
```

```python
dataset.head(10)
```

```python
mean_income = dataset.income.mean()
median_income = dataset.income.median()
mode_income = dataset.income.mode()

fig, ax = plt.subplots(1,4,figsize=(24,6))

sns.kdeplot(data=dataset, x="age", ax=ax[0])
sns.kdeplot(data=dataset, x="income", ax=ax[1])
sns.histplot(dataset, x="country", ax=ax[2])
sns.histplot(dataset, x="education", ax=ax[3])

ax[0].axvline(dataset.age.mean(),c=set2_orange, linewidth=5)
ax[0].axvline(dataset.age.median(),c=set2_blue, linewidth=5)
ax[0].axvline(dataset.age.quantile(q=.05),c=set2_rose, linewidth=2)
ax[0].axvline(dataset.age.quantile(q=.95),c=set2_rose, linewidth=2)
ax[0].hlines(y=.029, xmin=dataset.age.mean()-11.69, xmax=dataset.age.mean()+11.69, linewidth=2, color=set2_yellow)

ax[1].axvline(mean_income, c=set2_orange)
ax[1].axvline(median_income, c=set2_blue)

plt.show()
```

```python
plt.figure(figsize=(24,4))
sns.pointplot(x=dataset.education, y=dataset.accepted_cmp_total)

plt.show()
```

```python
dataset['marital_status'].replace(to_replace = ['Single'], value = 0, inplace = True)
dataset['marital_status'].replace(to_replace = ['Widow'], value = 1, inplace = True)
dataset['marital_status'].replace(to_replace = ['Divorced'], value = 2, inplace = True)
dataset['marital_status'].replace(to_replace = ['Together'], value = 3, inplace = True)
dataset['marital_status'].replace(to_replace = ['Married'], value = 4, inplace = True)

dataset['education'].replace(to_replace = ['Basic'], value = 0, inplace = True)
dataset['education'].replace(to_replace = ['Graduation'], value = 1, inplace = True)
dataset['education'].replace(to_replace = ['Master'], value = 2, inplace = True)
dataset['education'].replace(to_replace = ['PhD'], value = 3, inplace = True)
```

```python
dataset['country'].unique()
```

```python
dataset['country'].replace(to_replace = ['SP'], value = 0, inplace = True)
dataset['country'].replace(to_replace = ['CA'], value = 1, inplace = True)
dataset['country'].replace(to_replace = ['US'], value = 2, inplace = True)
dataset['country'].replace(to_replace = ['AUS'], value = 3, inplace = True)
dataset['country'].replace(to_replace = ['GER'], value = 4, inplace = True)
dataset['country'].replace(to_replace = ['IND'], value = 5, inplace = True)
dataset['country'].replace(to_replace = ['SA'], value = 6, inplace = True)
dataset['country'].replace(to_replace = ['ME'], value = 7, inplace = True)
```

```python
plt.figure(figsize=(28,8))
ax = sns.heatmap(dataset.corr(),annot=True,cmap='mako',cbar_kws={"shrink": .7, "label": "Correlation"}) 
ax.set_title("Correlation Matrix", fontsize=30) 
plt.show()
```

```python
fig, ax = plt.subplots(1,3, figsize=(24,6))

sns.boxplot(data=dataset, x="minors_home", y="income", ax=ax[0])
sns.boxplot(data=dataset, x="minors_home", y="purchases_total", ax=ax[1])
sns.boxplot(data=dataset, x="minors_home", y="mnt_total", ax=ax[2])

plt.show()
```

```python
mask = ['mnt_wines', 'mnt_fruits', 'mnt_meat_products', 'mnt_fish_products','mnt_sweet_products', 'mnt_gold_prods']

best_product_average = pd.DataFrame(round(dataset[mask].mean(),2).sort_values(),columns=['Average']).reset_index()
best_product_total = pd.DataFrame(round(dataset[mask].sum(),2).sort_values(),columns=['Spent']).reset_index()

fig, ax = plt.subplots(1,2, figsize=(24,6))
sns.barplot(x='Average', y='index', data=best_product_average,ax=ax[0])
ax[0].set_xlabel('Average Amount Spent')
ax[0].set_ylabel('Product Name')
ax[0].set_title('Average Amount Spent per category', size=20)

sns.barplot(x='Spent', y='index', data=best_product_total,ax=ax[1])
ax[1].set_xlabel('Total Amount Spent')
ax[1].set_ylabel('Product Name')
ax[1].set_title('Total Amount Spent per category', size=20)

fig.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(5,4))
dataset.groupby('country')['purchases_total'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Total Number of Purchases by Country', size=16)
plt.ylabel('Number of Purchases');

plt.figure(figsize=(5,4))
dataset.groupby('country')['mnt_total'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('Total Amount Spent by Country', size=16)
plt.ylabel('Amount Spent');
```

```python
cmp_success = pd.DataFrame(dataset[['accepted_cmp_1', 'accepted_cmp_2', 'accepted_cmp_3', 'accepted_cmp_4', 'accepted_cmp_5', 'response']].mean()*100, 
                           columns=['Percent']).reset_index()

sns.barplot(x='Percent', y='index', data=cmp_success.sort_values('Percent'), palette='Blues')
plt.xlabel('Accepted (%)')
plt.ylabel('Campaign')
plt.title('Marketing campaign success rate', size=16);
```
